
# vilion plot 

The unquestionable advantage of the violin plot over the box plot is that aside from showing the abovementioned statistics it also shows the entire distribution of the data. This is of interest, especially when dealing with multimodal data, i.e., a distribution with more than one peak.✨

refrence:https://towardsdatascience.com/violin-plots-explained-fb1d115e023d


```python

fig, ax = pyplot.subplots(figsize =(9, 7)) 
ax = sns.violinplot(ax=ax ,x="year", y="AveragePrice", hue="type", split=True, data=df, inner='quartile')
ax.set_title('Distribution of averge price per year', fontsize=16);
```

#  more visulaisation for reporting 

```python
#copied 
import plotly
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)


groupBy_whole = df.groupby('Date').mean()
scatter = go.Scatter(x=groupBy_whole.AveragePrice.index, y=groupBy_whole.AveragePrice)
data = [scatter]
layout=go.Layout(title="Time Series Plot for Mean Daily Price across all regions", xaxis={'title':'Date'}, yaxis={'title':'Prices'})
figure=go.Figure(data=data,layout=layout)
iplot(figure)
```
